```python
from matplotlib import pyplot
import numpy
```

```python
arr = numpy.load('../out.npz')
x = arr['energies']
ks = arr['ks']
```

```python
pyplot.plot(ks, x.mean(axis=0) / 8**4, 'x--')
```

```python
ks_flat = numpy.array([[k]*x.shape[0] for k in ks]).T.flatten()
xs_flat = x.flatten() / 8**4
pyplot.hist2d(ks_flat, xs_flat, bins=(len(ks), 256))
pyplot.show()
```
