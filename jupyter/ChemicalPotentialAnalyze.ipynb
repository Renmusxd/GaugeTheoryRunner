```python
import numpy
from matplotlib import pyplot
import matplotlib.cm as cm
```

```python
arr = numpy.load('data/L=8/out-k065-k085.npz')
arr._files

wm = arr["windings"].var(axis=0).mean(axis=-1)
pyplot.scatter(arr['replica_ks'], arr["replica_mus"], c=wm)
pyplot.colorbar()
pyplot.show()

rs_wm = wm.reshape(len(arr['mus']), len(arr['ks']))
pyplot.plot(arr['ks'], rs_wm[rs_wm.shape[0]//2,:], 'x-')
pyplot.grid()
pyplot.show()

wm = arr["windings"].mean(axis=0).mean(axis=-1)
pyplot.scatter(arr['replica_ks'], arr["replica_mus"], c=wm)
pyplot.colorbar()
pyplot.show()
```

```python
for L in [4,6,8,10,12,14,16]:
    arr = numpy.load(f'data/L={L}/out-k065-k085.npz')
    wm = arr["windings"].var(axis=0).mean(axis=-1)
    rs_wm = wm.reshape(len(arr['mus']), len(arr['ks']))
    pyplot.plot(arr['ks'], rs_wm[rs_wm.shape[0]//2,:], 'x-', label=str(L))

pyplot.legend()
pyplot.grid()
pyplot.show()
```
